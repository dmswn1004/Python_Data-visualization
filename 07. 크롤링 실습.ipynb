## 여러 페이지 데이터 추출

```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

### 페이지 번호 변경

```python
browser = webdriver.Chrome("./chromedriver")

for i in range(1,6):
    url = "https://section.blog.naver.com/Search/Post.naver?pageNo="+str(i)+"&rangeType=ALL&orderBy=sim&keyword=%EB%B2%9A%EA%BD%83"
    browser.get(url)
    time.sleep(1)
    search = browser.find_elements(By.CSS_SELECTOR, "div.desc")
    for j in search:
        title = j.find_element(By.CSS_SELECTOR, "span.title").text
        desc = j.find_element(By.CSS_SELECTOR, "a.text").text
        print(title + " - " + desc)

browser.quit()
```

### 페이지 탐색 스킬

```python
browser = webdriver.Chrome("./chromedriver")
url = "https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EC%A0%95%EC%99%95%EB%8F%99+%EC%84%B8%EA%B3%84%EB%A1%9C%EB%A7%88%ED%8A%B8&nso=so%3Ar%2Cp%3Aall%2Ca%3Aall"
browser.get(url)
time.sleep(1)

while 1:
    titles = browser.find_elements(By.CSS_SELECTOR, "a.news_tit")
    for i in titles:
        print(i.text)
    try:
        browser.find_element(By.CSS_SELECTOR, "a.btn_next").click()
        time.sleep(1)
    except:
        print("데이터 수집 완료")
        break

browser.quit()
```

### 네이버 뉴스 크롤러

```python
browser = webdriver.Chrome("./chromedriver")
url = "https://search.naver.com/search.naver?where=news&sm=tab_jum&query=산불"
browser.get(url)
time.sleep(1)

news_links = browser.find_elements(By.CSS_SELECTOR, "div.info_group > a.info:not(.press)")
for i in news_links:
    i.click()
    browser.switch_to.window(browser.window_handles[1]) # 2번째 탭으로 변경 
    time.sleep(1)
    title = browser.find_element(By.CSS_SELECTOR, "#title_area > span").text 
    date = browser.find_element(By.CSS_SELECTOR, "div.media_end_head_info_datestamp > div:nth-child(1) > span").text
#     article = browser.find_element(By.CSS_SELECTOR, "#dic_area").text 
    print('//제목:', title, '\n//시간:', date, '\n\n')
    browser.close() # 탭 닫기 
    browser.switch_to.window(browser.window_handles[0]) # 1번째 탭으로 복귀
browser.quit()
```

### 다음 뉴스 크롤러 만들기

```python
browser = webdriver.Chrome("./chromedriver")
for i in range(1,4):
    url = "https://search.daum.net/search?w=news&DA=PGD&enc=utf8&cluster=y&cluster_page=1&q=%ED%95%9C%EA%B5%AD%EA%B3%B5%ED%95%99%EB%8C%80%ED%95%99%EA%B5%90&p="+str(i)
    browser.get(url)
    time.sleep(1)
    daum_news = browser.find_elements(By.CSS_SELECTOR,"span.cont_info > a.f_nb:nth-child(3)")
    for j in daum_news:
        j.click()
        browser.switch_to.window(browser.window_handles[1]) # 2번째 탭으로 변경 
        time.sleep(1)
        title = browser.find_element(By.CSS_SELECTOR, "h3.tit_view").text
        date = browser.find_element(By.CSS_SELECTOR, "span.num_date").text
        print('//제목:', title, '  -  //시간:', date, '\n')
        browser.close() # 탭 닫기 
        browser.switch_to.window(browser.window_handles[0]) # 1번째 탭으로 복귀
browser.quit()
```
