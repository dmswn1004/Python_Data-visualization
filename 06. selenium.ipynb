## 뉴스 검색

```python
browser = webdriver.Chrome("./chromedriver")
url = "https://search.naver.com/search.naver?where=news&sm=tab_jum&query=한국공학대"
browser.get(url)
time.sleep(1)
news_list = browser.find_elements(By.CSS_SELECTOR, ".news_area")
for i in news_list:
    sou = i.find_element(By.CSS_SELECTOR, ".info.press").text
    tit = i.find_element(By.CSS_SELECTOR,".news_tit").text
    print(sou + " - " + tit)
```

## 멜론 차트

```python
browser = webdriver.Chrome("./chromedriver")
url = "https://www.melon.com/chart/index.htm"
browser.get(url)
time.sleep(1)

music_list = browser.find_elements(By.CSS_SELECTOR, "#lst50")
for i in music_list:
    ranking = i.find_element(By.CSS_SELECTOR, "span.rank").text
    title = i.find_element(By.CSS_SELECTOR, "div.ellipsis.rank01").text
    singer = i.find_element(By.CSS_SELECTOR, "div.ellipsis.rank02").text
    print(ranking + "위 " + singer + " - " + title)
```

## 네이버 부동산

```python
browser = webdriver.Chrome("./chromedriver")
url = "https://new.land.naver.com/complexes/110814?ms=37.3593216,126.7176545,NaN&a=APT:ABYG:JGC&e=RETAIL"
browser.get(url)
time.sleep(1)

apt_list = browser.find_elements(By.CSS_SELECTOR, "a.item_link")
for i in apt_list:
    size = i.find_element(By.CSS_SELECTOR, "span.spec").text
    price = i.find_element(By.CSS_SELECTOR, "div.price_line").text
    print(size + " - " + price)
```
