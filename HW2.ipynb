## (문제)
IMDb는 아마존의 자회사인 영화 정보 사이트로, 평점이 가장 높은 Top 250 영화를 확인할 수 있습니다.

## No. 1 (1점)
**BeautifulSoup**을 이용하여 다음 url에서 제목, 연도, 평점 정보를 수집하세요.

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.imdb.com/chart/top/'
headers = {'User-Agent': 'Mozilla/5.0'} 
response = requests.get(url, headers = headers)
```

```python
soup = BeautifulSoup(response.content, "html.parser", from_encoding="utf-8")
movies_list = soup.select("tbody.lister-list tr")
for i in movies_list:
    title = i.select_one("td.titleColumn > a").text
    year = i.select_one("span.secondaryInfo").text[1:5] # 괄호 없애기 위해 슬라이싱해주었습니다.
    rating = i.select_one(".imdbRating strong").text
    print("Title: " + title + " / Year: " + year + " / Rating: " + rating)
```

### 문제 설명
IMDb Top 250 웹페이지의 URL을 지정하고 사용자 에이전트를 모질라 브라우저로 지정하는 headers 매개변수를 설정합니다.  
그런 다음 requests.get() 함수를 사용하여 URL과 headers를 인수로 전달하여 웹페이지에 GET 요청을 보냅니다.

BeautifulSoup의 select() 메소드를 사용하여 HTML 콘텐츠에서 모든 영화 목록을 선택하고 각 항목을 루프를 통해 반복합니다.  
for 루프에서는 moves_list 목록에서 title 변수에는 제목과 year 변수에는 연도를, rating 변수에는 평점 정보를 추출해 넣어주었습니다.  

## No. 2 (2점)
수집한 정보들을 데이터프레임 형태로 변환하세요.  
해당 데이터프레임에는 순위('Rank'), 제목('Title'), 연도('Year'), 평점('Rate') 열이 존재해야 합니다.  
특히 'Rank'와 'Year'는 int, 'Rate'는 float 자료형으로 변환하세요.  

```python
data = []
movies_list = soup.select("tbody.lister-list tr")
for i,j in zip(movies_list,range(len(movies_list)+1)):
    rank = j + 1
    title = i.select_one("td.titleColumn > a").text
    year = int(i.select_one("span.secondaryInfo").text[1:5]) # 괄호 없애기 위해 슬라이싱해주었습니다.
    rating = float(i.select_one(".imdbRating strong").text)
    print(f"Title: {title} / Year: {year} / Rating: {rating}")
    data.append((rank, title, year, rating))
```

```python
import pandas as pd
df = pd.DataFrame(data, columns=['Rank', 'Title', 'Year', 'Rating'])
df.info()
```

```python
df2 = df[:]
df2 = df2.astype({'Rating':'float'}) # 이 방식으로도 데이터 자료형 변경 가능
```

data 라는 리스트를 생성하고 각 변수의 값을 append를 통해 넣어주었습니다.  

pandas 패키지를 사용하여 data 리스트의 값을 데이터프레임으로 만들어주고 열 값을 ['Rank', 'Title', 'Year', 'Rating']으로 지정하였습니다.

## No. 3 (2점)
해당 데이터를 10년 단위로 정리하겠습니다.  
데이터가 1920년대부터 존재하므로, 1920년대부터 2020년대까지 총 11개로 나눌 수 있습니다.  
Top 250을 대상으로, 시대별로 (1) Top 250에 포함된 영화 수, (2) 평균 평점, (3) 최고 평점 영화를 출력하세요.  
최고 평점 영화는 시대별로 가장 순위가 높은 영화를 출력하시면 됩니다.  


ex) 출력 결과:
1920년대 - 영화 수: 6 / 평균 평점: 8.13 / 최고 평점 영화: Metropolis  
1930년대 - 영화 수: 7 / 평균 평점: 8.17 / 최고 평점 영화: 모던 타임즈  
.....  
2020년대 - 영화 수: 6 / 평균 평점: 8.13 / 최고 평점 영화: Hamilton  

```python
df['Decade'] = (df['Year'] // 10) * 10 
df.head()
```

### 연습

```python
grouped = df.groupby("Decade")
for year, group in grouped:
    num_movies = len(group)
    avg_rating = group["Rating"].mean()
    highest_rated = group["Rating"].max()
    print(f"{year} - 영화 수: {num_movies} / 평균 평점: {avg_rating:.2f} / 최고 평점 영화: {highest_rated}")
```

```python
decade_stats =  df.groupby('Decade').agg({'Title': 'count', 'Rating': ['mean', 'max']})
decade_stats.columns = ['영화 수', '평균 평점', '최고 평점 영화']
decade_stats['평균 평점'] = decade_stats['평균 평점'].round(2)
decade_stats
```

```python
for values, i in decade_stats.items():
    count = decade_stats['영화 수'].values
    avg_rating = decade_stats['평균 평점'].values
    max_rating = decade_stats['최고 평점 영화'].values
    
for i in range(len(count)):  
    print(f"영화 수: {count[i]} / 평균 평점: {avg_rating[i]} / 최고 평점 영화: {max_rating[i]}")
```

각 영화의 Year의 값을 10년 단위로 변환한 값을 Decade라는 열에 추가했습니다.  

각 연도에 대한 영화수 평균 평점, 최고 평점 영화를 구해줍니다.  
영화수에 대한 평균 평점의 소수점을 2자리로 지정해줍니다.  

각각의 데이터프레임에 있는 값을 각각의 리스트에 저장한 후 리스트의 인덱스를 통해 출력해줍니다.  
for문을 통해 데이터를 출력해줍나다.
