## 실습 1 - 아이폰 모델명 추출

```python
import requests

url = "https://market.cetizen.com/market.php?q=view&auc_no=24297415"
headers = {'User-Agent': 'Mozilla/5.0'} 
response = requests.get(url, headers = headers)
print(response.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser", from_encoding="utf-8")
name = soup.select("span.fb.fsize7")
print(name[0].text)
for i in name:
    print(i.text)
```

```python
name_list = soup.select("span.fb")
for i in name_list:
    print(i.text)
```

## 실습 3 - 네이버 뉴스

```python
url = "https://search.naver.com/search.naver?where=news&sm=tab_jum&query=한국공학대"
headers = {'User-Agent': 'Mozilla/5.0'} 
response = requests.get(url, headers = headers)
print(response.text)
```

```python
soup = BeautifulSoup(response.content, "html.parser", from_encoding="utf-8")
news_info = soup.select("a.info.press")
for i in news_info:
    print(i.text)
```

```python
news_tit = soup.select("a.news_tit")
for i in news_tit:
    print(i.text)
```

```python
for i, j in zip(news_info, news_tit):
    print(i.text + " - " + j.text)
```

```python
news_list = soup.select(".news_area")
for i in news_list:
    print(i.select_one("a.info.press").text + " - " + i.select_one("a.news_tit").text)
```
