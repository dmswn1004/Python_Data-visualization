## 삼성전자 PER 추출

```python
url = "https://finance.naver.com/item/main.nhn?code=005930"
headers = {'User-Agent': 'Mozilla/5.0'} 
response = requests.get(url, headers = headers)
print(response.text)
```

```python
soup = BeautifulSoup(response.content, "html.parser", from_encoding="utf-8") 
per_list = soup.select("#_per")
print(per_list)
```

```python
per = soup.select_one("#_per")
print(per)
```

```python
print(per.text)
```

```python
dvr = soup.select_one("#_dvr")
print(dvr.text)
```

## 학교 홈페이지 추출

```python
url = "https://www.tukorea.ac.kr/tukorea/3475/subview.do"
headers = {'User-Agent': 'Mozilla/5.0'} 
response = requests.get(url, headers = headers)
print(response.text)
```

```python
soup = BeautifulSoup(response.content, "html.parser", from_encoding="utf-8") 
dept = soup.select("p.p_1 strong")
# print(dept)
for i in dept:
    print(i.text)
```

```python
dept_eng_list = soup.select("p.p_1 span")
for i in dept_eng_list:
    print(i.text)
```

```python
dept_des = soup.select("p.p_2")
for i in dept_des:
    print(i.text)
```

```python
for i, j, k in zip(dept, dept_eng_list, dept_des):
    print(i.text + " - " + j.text + " - " + k.text)
```

```python
dept_all_list = soup.select("div.con")

for i in dept_all_list:
    print(i.select_one("p.p_1 strong").text + " - " + i.select_one("p.p_1 span").text + " - " + i.select_one("p.p_2").text)
```
