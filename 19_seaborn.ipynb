```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
```

```python
type(iris)
```

```python
iris.head()
```

```python
x = iris.petal_length.values

sns.rugplot(x)
plt.rcParams['figure.figsize'] = (4,3)
plt.show()
```

```python
sns.kdeplot(x)
plt.show()
```

```python
sns.displot(x, kde=True, rug=True)
plt.show()
```

```python
sns.countplot(x='species', data=iris)
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris)
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="kde")
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="hist")
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, hue="species")
plt.show()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris, hue='species')
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
titanic_size = titanic.pivot_table(index="class", columns="sex", aggfunc="size")
titanic_size
```

```python
sns.heatmap(titanic_size, cmap=sns.light_palette('gray', as_cmap=True), annot=True, fmt='d')
plt.show()
```

```python
sns.barplot(x="species", y="sepal_width", data=iris)
plt.show()
```

```python
sns.boxplot(x="species", y="sepal_width", data=iris)
plt.show()
```

```python
sns.violinplot(x="species", y="sepal_width", data=iris)
plt.show()
```

```python
sns.stripplot(x="species", y="sepal_width", data=iris)
plt.show()
```

```python
sns.swarmplot(x="species", y="sepal_width", data=iris)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
sns.barplot(x="pclass", y="survived", hue='sex', data=titanic)
plt.show()
```

```python
sns.boxplot(x="pclass", y="age", hue='sex', data=titanic)
plt.show()
```

```python
sns.violinplot(x="pclass", y="age", hue='sex', data=titanic)
plt.show()
```

```python
sns.violinplot(x="pclass", y="age", hue='sex', data=titanic, split=True)
plt.show()
```

```python
sns.stripplot(x="pclass", y="age", hue='sex', data=titanic, dodge=True)
plt.show()
```

```python
sns.catplot(x="age", y="sex", hue='survived', row="class", data=titanic, kind="violin", split=True, heigth=2, aspect=4)
plt.show()
```

```python
sns.boxplot(x="species", y="sepal_width", data=iris)
sns.stripplot(x="species", y="sepal_width", data=iris, color='0.8')
plt.show()
```

```python
sns.violinplot(x="species", y="sepal_width", data=iris)
sns.swarmplot(x="species", y="sepal_width", data=iris, color='0.8')
plt.show()
```
