## Quiz - 연습문제 4.4.3

### 1. 타이타닉호 승객의 평균 나이를 구하라.

```python
df['age'].mean()
```

### 2. 타이타닉호 승객중 여성 승객의 평균 나이를 구하라.

```python
df.loc[df['sex']=='female']['age'].mean()
```

### 3. 타이타닉호 승객중 1등식 선실의 여성 승객의 평균 나이를 구하라.

```python
df.loc[(df['sex']=='female') & (df['pclass']==1)].age.mean()
```

## apply 변환

```python
df['adult'] = df.apply(lambda x: True if x.age >= 20 else False, axis = 1)
df
```

## Quiz - 연습문제 4.4.4

### 1. 20살이 넘으면 성별을 그대로 사용한다. 20살 미만이면 성별에 관계없이 "child"라고 한다.

```python
df['who'] = df.apply(lambda x: x.who if x.age >= 20 else 'child', axis = 1)
df[['age','who']]
```

### fillna() 
: NaN값(결측치)을 원하는 값으로 채우기  
- 예시) 평균값으로 채우기

```python
df2 = df[:]
df2['age'] = df2['age'].fillna(round(df2['age'].mean(),1))
df2
```

```python
df2 = df2.astype({'pclass':'float'})
df2.info()
```
